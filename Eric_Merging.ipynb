```python
import pandas as pd
```

```python
y1213 = pd.read_csv("1213.csv",dtype = {0 : "string"})
y1420 = pd.read_csv("1420.csv",dtype = {0 : "string"})
y2021 = pd.read_excel("2021.xlsx",dtype = {0 : "string"})
```

```python
y2022 = pd.read_excel("2022.xlsx",dtype = {0 : "string"})
y2023 = pd.read_csv("2023.csv",dtype = {0 : "string"})
y2024 = pd.read_excel("2024.xlsx",dtype = {0 : "string"})
```

```python
y1213f = y1213.columns.tolist()
y1420f = y1420.columns.tolist()
y2021f = y2021.columns.tolist()
y2022f = y2022.columns.tolist()
y2023f = y2023.columns.tolist()
y2024f = y2024.columns.tolist()
yflist = [y1213f,y1420f,y2021f,y2022f,y2023f,y2024f]
```

```python
def check(l):
    isid = 1
    names = ["1213", "1420", "2021", "2022", "2023", "2024"]
    for i in range (len(l)-1):
        base = l[i]
        for j in range (i+1 , len(l)):
            if l[j] != base:
                isid = 0
                print(f"{names[i]} != {names[j]}")
    if isid == 1:
        print("all equal")

    
```

```python
check(yflist)
```

```python
unmatched_2024 = sorted(set(y2024f) - set(y1213f))
unmatched_base = sorted(set(y1213f) - set(y2024f))
print(unmatched_2024, unmatched_base)
```

```python
import difflib

base_cols = y1213.columns.tolist()
y24_cols  = y2024.columns.tolist()

base_set = set(base_cols)

rn_map = {}

for f in y24_cols:
    if f in base_set:
        continue 
    matches = difflib.get_close_matches(f, base_cols, n=1, cutoff = 0.6)
    if matches:
        rn_map[f] = matches[0]


```

```python
rename_map = {
    'MaxInjuryS': 'MaxInjurySeverity',
    'CrashSever': 'CrashSeverity',
    'RoadwayAcc': 'RoadwayAccessControlCde',
    'LightCondi': 'LightCondition',
    'RoadwayCha': 'RoadwayCharacteristic',
    'RoadSurfac': 'RoadSurfaceCondition',
    'TrafficCon': 'TrafficControl',
    'WeatherCon': 'WeatherCondition',
    'Commercial': 'CommercialVehicleCrashInd',
    'DMVInsertD': 'DMVInsertDate',
    'NumberOfFa': 'NumberOfFatalities',
    'NumberOfIn': 'NumberOfInjuries',
    'NumberOfOt': 'NumberOfOtherInjuries',
    'NumberOfSe': 'NumberOfSeriousInjuries',
    'NumberOfVe': 'NumberOfVehicles',
    'NonReporta': 'NonReportable',
    'ReportingA': 'ReportingAgency',
    'NonPublicW': 'NonPublicWayCode',
    'ClosestCro': 'ClosestCrossStreet',
    'MasterInte': 'MasterIntersectionId',
    'Intersecti': 'IntersectionIndicator',
    'CityTownNa': 'CityTownName',
    'ReferenceM': 'ReferenceMarker',
    'MAINT_JURI': 'MAINT_JURISDICTION_TYPE_ID',
    'OWNING_JUR': 'OWNING_JURISDICTION_TYPE_ID',
    'DirectionF': 'DirectionFromIntersection',
    'DistanceFr': 'DistanceFromIntersection'
}
```

```python
y2024_new = y2024.copy().rename(columns = rename_map)
y2024f_new = y2024_new.columns.tolist()
```

```python
yflist2 = [y1213f,y1420f,y2021f,y2022f,y2023f,y2024f_new]
check(yflist2)
```

```python
unmatched_2024_new = sorted(set(y2024f_new) - set(y1213f))
unmatched_base = sorted(set(y1213f) - set(y2024f_new))
print(unmatched_2024_new, unmatched_base)
```

```python
drop_2024 = ['Extra columns', 'Lat', 'Long', 'OBJECTID', 'Unnamed: 47', 'Unnamed: 48', 'Unnamed: 49','CollisionT']
y2024_new_d = y2024_new.copy().drop(columns = drop_2024)
```

```python
y2024_new_d["CrashDate"] = pd.to_datetime(y2024_new_d["CrashDate"], errors="coerce").dt.date
```

```python
dflist = [y1213,y1420,y2021,y2022,y2023,y2024_new_d]
result = pd.concat(dflist, ignore_index=True)
```

```python
result
```

```python
result = result.drop(columns = ['CrashTime','CrashTypeCde','ApparentFactorCodes'])
```

```python
result["CrashDate"] = pd.to_datetime(result["CrashDate"], errors="coerce").dt.date
```

```python
result["IntersectionIndicator"] = (
    pd.to_numeric(result["IntersectionIndicator"], errors="coerce")
    .astype("Int64") 
    .astype("string")
)
```

```python
na_tokens = ["", " ", "  ", "\t", "\n", "NULL", "null", "NaN", "nan", "N/A", "n/a", "NA", "na"]
result = result.replace(na_tokens, pd.NA)
```

```python
col_s = result.select_dtypes(include=["object", "string"]).columns

for col in col_s:
    result[col] = (result[col].astype("string").str.strip().str.replace(r"\s+", " ", regex = True).str.title())
```
